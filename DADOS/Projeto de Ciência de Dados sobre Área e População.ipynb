```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
seaborn.set()
```

```python
cities = pd.read_csv("california_cities.csv")
cities.head()
```

```python
# Extração da área de interesse
latitude, longitude = cities['latd'], cities['longd']
population, area = cities['population_total'], cities['area_total_km2']
```

```python
#Espalhar os pontos usando tamanho e cor, sem a parte do rótulo
plt.scatter(longitude, latitude, label=None, c=np.log10(population),
           cmap='viridis', s=area, linewidths=0, alpha=0.5)
plt.axis('equal')
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.colorbar(label='log$_{10}$(population)')
plt.clim(3,7)

#Criar a legenda, plotar as listas vazis com o tamanho e rótulo
for area in [100, 300, 500]:
    plt.scatter([], [], c='k', alpha=0.3, s=area, label=str(area) + 'km$^2$')
plt.legend(scatterpoints=1, frameon=False, labelspacing=1, title='City Areas')
plt.title("Area and Population of California Cities")
plt.show()
```
